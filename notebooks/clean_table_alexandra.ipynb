```python
import pandas as pd 
```

```python
doc1 = 'C:/Users/Alexandra/Documents/Ironhack/Week4/Project/first_project/data/raw/train/df_Customers.csv'
customer_df = pd.read_csv(doc1)
customer_df.dropna(how='all', inplace=True)
customer_df.drop_duplicates(inplace=True)
customer_df
```

```python
doc2 = 'C:/Users/Alexandra/Documents/Ironhack/Week4/Project/first_project/data/raw/train/df_OrderItems.csv'
order_items_df = pd.read_csv(doc2)
order_items_df.dropna(how='all', inplace=True)
order_items_df.drop_duplicates(inplace=True)
order_items_df
```

```python
doc3 = 'C:/Users/Alexandra/Documents/Ironhack/Week4/Project/first_project/data/raw/train/df_Orders.csv'
orders_df = pd.read_csv(doc3)
orders_df.dropna(how='all', inplace=True)
orders_df.drop_duplicates(inplace=True)
orders_df
```

```python
doc4 ='C:/Users/Alexandra/Documents/Ironhack/Week4/Project/first_project/data/raw/train/df_Payments.csv'
payments_df = pd.read_csv(doc4)
payments_df.dropna(how='all', inplace=True)
payments_df.drop_duplicates(inplace=True)
payments_df
```

```python
payments_df = payments_df.reset_index().rename(columns={'index': 'payment_id'})
payments_df["payment_id"] = payments_df["payment_id"] +1
```

```python
display(payments_df)
```

```python
doc5 = 'C:/Users/Alexandra/Documents/Ironhack/Week4/Project/first_project/data/raw/train/df_Products.csv'
products_df = pd.read_csv(doc5)
products_df.dropna(how='all', inplace=True)
products_df.drop_duplicates(inplace=True)
products_df
```
