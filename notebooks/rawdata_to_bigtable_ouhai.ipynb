```python
import pandas as pd
```

```python
customers = pd.read_csv("../data/raw/df_Customers.csv")
orders = pd.read_csv("../data/raw/df_Orders.csv")
order_items = pd.read_csv("../data/raw/df_OrderItems.csv")
products = pd.read_csv("../data/raw/df_Products.csv")
payments = pd.read_csv("../data/raw/df_Payments.csv")
```

```python
payments = payments.copy()
payments.insert(0, "payment_id", range(1, len(payments) + 1))
```

```python
payments['payment_id'].head()
```

```python
df = orders.merge(customers, on="customer_id", how="left")
df = df.merge(order_items, on="order_id", how="left")
df = df.merge(payments, on="order_id", how="left")
df = df.merge(products, on="product_id", how="left")
```

```python
df=df.dropna()
```

```python
df = df.drop_duplicates()
```
